How does LLM works?
Large Language Model (LLM) is a type of artificial intelligence (AI) that is trained on large amounts of text data to generate human-like language. LLMs are designed to learn patterns and relationships in language, and they can be used for a variety of tasks such as language translation, text summarization, and question answering.

What is the Hallucination?
Hallucination is a phenomenon in which a model generates text that is not supported by the input data . This can occur when the model is overconfident in its predictions or when it is trying to
generate text that is not supported by the input data.

What is RAG (Retrieval Augmented Generaltion) ?
RAG is a type of AI model that combines the strengths of retrieval-based and generative models.
