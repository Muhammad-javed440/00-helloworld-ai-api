```python
from agents import enable_verbose_stdout_logging, function_tool
enable_verbose_stdout_logging()
@function_tool
def get_weather(location: str)->str:
  """Return Weather for the given location
  Args:
    location: Location name

    Returns:
    weather
  """
  return f"The weather in {location} is sunny"

@function_tool
def get_current_news(industry:str)->str:
  """Return the current news for the given industry
  Args:
    industry: industry to look in current news

    Returns:
    news
  """
  return f"The current news in {industry} is news"

agent: Agent = Agent(
    name="weather agent",
    instructions="You provide weather information",
    model=model,
    tools=[ get_weather,get_current_news ]
)

agent_response = Runner.run_sync(
        starting_agent=agent,
        input="tell me about weather of lahore,pakistan"
    )
print(agent_response.final_output)
```
