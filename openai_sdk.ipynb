# Run google gemini with openai-agent sdk

```python
import os
from google.colab import userdata
from agents import Agent, Runner, AsyncOpenAI, OpenAIChatCompletionsModel
from agents.run import RunConfig
from openai.types.responses import ResponseTextDeltaEvent
import asyncio
```

```python
gemini_api_key = userdata.get("GEMINI_API_KEY")

# Check if the api key is present; if not , raise an error
if not gemini_api_key:
    raise ValueError("GEMINI_API_KEY not found in userdata")
# Reference: https://ai.google.dev/gemini-api/docs/openai
external_client=AsyncOpenAI(
    api_key=gemini_api_key,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/",
)

model=OpenAIChatCompletionsModel(
     model="gemini-2.0-flash",
     openai_client=external_client
 )

config= RunConfig(
    model=model,
    model_provider=external_client,
    tracing_disabled=True
)
```

# Hello World code | method one

```python
async def main():
  agent=Agent(
      name="Joker",
      model=model,
      instructions="You are a helpful assistant.",
  )
async def main():
  agent=Agent(
      name="Joker",
      model=model,
      instructions="You are a helpful assistant.",
  )
  result = Runner.run_streamed(agent, input="Please tell me five jokes and translate it in urdu")
  async for event in result.stream_events():
    if event.type == "raw_response_event" and isinstance(event.data, ResponseTextDeltaEvent):
      print(event.data.delta, end="", flush=True)

asyncio.run(main())
```

```python
import asyncio
from agents import Agent, Runner

async def main():
   agent=Agent(
      name="Assistant",
      model=model,
      instructions="You only respond in haikus."
   )
   # run_streamed returns a RunResultStreaming object, not an awaitable
   # it can be used in a for loop to get responses as they are produced.
   result = Runner.run_streamed(agent, "Tell me about recursion in programming.", run_config=config)
   # Instead of awaiting the function, loop through response events:
   async for event in result.stream_events():
       print(event)

asyncio.run(main())
```
