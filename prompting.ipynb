# What is a chatbot:

A chatbot is a software application designed to simulate human-like conversations with users. Chatbots can interact with people through text or voice and are commonly used in various applications to provide support, answer questions, or perform tasks.

```python
import PIL.Image        # The PIL module, which stands for Python Imaging Library
img = PIL.Image.open('/content/background.jpeg')
img
```

```python
prompt = """This image contains a sketch of a potential product along with some notes.
Given the product sketch, describe the product as thoroughly as possible based on what you
see in the image, making sure to note all of the product features. Return output in json format:
{description: description, features: [feature1, feature2, feature3, etc]}"""
```

```python
model = genai.GenerativeModel('gemini-2.0-flash-exp')
response = model.generate_content([prompt, img])
Markdown(response.text)
```

```python
model = genai.GenerativeModel('gemini-2.0-flash-exp')
chat = model.start_chat(history=[])      # To store chat in his memory in the form of oject of array
```

```python
response = chat.send_message("In one sentence, explain how a computer works to a young child.")
Markdown(response.text)
```

```python
response = chat.send_message("Okay, how about a more detailed explanation to a high schooler?")
print(response.text)
```

```python
print(chat.history)
```

```python
model = genai.GenerativeModel(
    'gemini-2.0-flash-exp',
    generation_config=genai.GenerationConfig(
        max_output_tokens=2000,
        temperature=0.9,
    ))
```

```python
response = model.generate_content(
    'Give me a numbered list of cat facts. Limit to 5 facts', # Combine the prompt and the additional string
    generation_config=genai.GenerationConfig(stop_sequences=['\n6'])
)
```

```python
print(response.text)
```
